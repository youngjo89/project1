```python
def TM_transform(series, T_replace, M_replace): 
    """
    데이터내의 T, M을 원하는 값으로 바꿔주는 함수
    TM_transform(series, T_replace, M_replace)
    """
    series = series.astype(str).map(lambda s: s.strip())
    series[series == 'T'] = T_replace
    series[series =='M'] = M_replace
    return series.astype('float')
```

```python
weather_test = weather_new.drop(['sunrise', 'sunset', 'codesum'], axis = 1)
weather_test_station1 = weather_test[weather_test['station_nbr'] == 1]
weather_test_station1.reset_index(drop = True, inplace = True)
weather_test_station1['depart'] = TM_transform(weather_test_station1['depart'], 0, 0) #depart
weather_test_station1['dewpoint'] = TM_transform(weather_test_station1['dewpoint'], 0, 0) #dewpoint
```
